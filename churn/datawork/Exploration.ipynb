```python
trans = pd.read_csv(PATH/'transactions.csv', parse_dates = ['membership_expire_date', 'transaction_date'])
trans.head()
```

```python
trans['month'] = trans['transaction_date'].dt.month
trans['year'] = trans['transaction_date'].dt.year

trans.groupby(['year', 'month'])['actual_amount_paid'].sum()
```

```python
trans.groupby()
```

```python
rev = trans[(trans['membership_expire_date'] < trans['transaction_date']) | 
            ((trans['is_cancel'] == 0) & (trans['membership_expire_date'] == trans['transaction_date']))].copy()
rev.shape
```

```python
rev_members = rev['msno'].unique()
len(rev_members)
```

```python
n_members = trans['msno'].nunique()
n_members
```

```python
trans.shape
```

```python
trans[~trans['msno'].isin(rev_members)].shape
```

```python
to_exclude = []

for i, (msno, grouped) in enumerate(trans.groupby('msno')):
    rev = grouped[(grouped['membership_expire_date'] < grouped['transaction_date']) | 
            ((grouped['is_cancel'] == 0) & (grouped['membership_expire_date'] == grouped['transaction_date']))]
    if len(rev) > 0:
        to_exclude.append(msno)
        
    print(f'{round((100 * (i / n_members)), 2)}% complete.', end = '\r')
```
